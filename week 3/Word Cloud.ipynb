```python
# importing libraries
from wordcloud import WordCloud, STOPWORDS
```

```python
# download file and save as alice_novel.txt
#!wget --quiet https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/alice_novel.txt

# open file and read the data
alice_novel = open('alice_novel.txt','r').read()

print('File downloaded and saved!')
```

We use the function `set()` to remove any redundant stopwords.

```python
stopwords = set(STOPWORDS)
```

```python
# create a wordcloud from the first 2000 words from the text file

# instantiate an object
alice_wc = WordCloud(background_color='white',
                    max_words= 2000,
                    stopwords= stopwords)

# generate world coloud 
alice_wc.generate(alice_novel)
```

## Importing Libraries

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

## Plot the image

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Lets change the figure size

```python
fig = plt.figure()
fig.set_figheight(18)
fig.set_figwidth(14)

# display the word cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# add the word said to stopwords
stopwords.add('said')

# generate the word cloud again
alice_wc.generate(alice_novel)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

Excellent! This looks really interesting! Another cool thing you can implement with the `word_cloud` package is superimposing the words onto a mask of any shape. Let's use a mask of Alice and her rabbit. We already created the mask for you, so let's go ahead and download it and call it *alice_mask.png*.

```python
from PIL import Image
import numpy as np 

alice_mask = np.array(Image.open('alice_mask.png'))
print('Image imported!')
```

```python
# view the image
fig = plt.figure()
fig.set_figheight(18)
fig.set_figwidth(14)

plt.imshow(alice_mask, cmap = plt.cm.gray, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a new wordcloud with mask
alice_wc = WordCloud(background_color='white',
                    max_words= 200,
                    mask=alice_mask,
                    stopwords=stopwords)

alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
