```python
import pandas as pd
```

```python
df_can = pd.read_excel('datasets_Canada.xlsx',
                      sheet_name='Canada by Citizenship',
                      skiprows=(20),
                      skipfooter=2)
df_can.head(2)
```

```python
df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DevName'], axis = 1, inplace=True)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName': 'Continent', 'Regname': 'Region'}, inplace = True)
```

```python
df_can.head(2)
```

```python
years = list(map(str, range(1980,2014)))
print(years)
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
df_new = pd.DataFrame(df_can[years].sum(axis = 0))  # o means column wise
```

```python
df_new.head(3)
```

```python
# convert the index column from string to integer
df_new.index = df_new.index.map(int)

# only after that reset the index column
df_new.reset_index(inplace=True)
```

```python
df_new.columns = ['year', 'total']
df_new.head()
```

# Regression Plots

```python
import seaborn as sns
```

```python
ax = sns.regplot(x='year',
                 y='total',
                 data=df_new)

ax.figure.set_size_inches(10,6)
```

It contains also 95% confidence interval

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.figure(figsize=(12,7)) # change the figure size

sns.set(font_scale=1.2)    # change the fontsize of everything in the figure

ax = sns.regplot(x='year', y='total',
                data= df_new,
                color = 'g',         # change the marker color
                marker = '+',        # change the marker shape
                scatter_kws={'s':100} # change the marker size
                )

ax.set(xlabel = 'Year', ylabel = 'Total Immigration')       # set xlabel and ylabel
ax.set_title('Total Immigration to Canada from 1980-2013')  # set the title
```

**We can remove the background also.**

```python
plt.figure(figsize = (12,7))
sns.set_style('ticks')  # set the background to white
                        # use 'whitegrid' for white backgrounds with gridlines

ax = sns.regplot(x='year', y='total',
                data= df_new,
                color = 'g',         # change the marker color
                marker = '+',        # change the marker shape
                scatter_kws={'s':100} # change the marker size
                )
```
